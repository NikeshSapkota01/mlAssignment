```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('combined_data.csv')
df.head()
```

```python
df.index = pd.to_datetime(df.index)
df.index = df.index.tz_localize(None)
```

```python
# Creating a new feature for better representing day-wise values
df['Mean'] = (df['low'] + df['high'])/2
```

```python
# Cleaning the data for any NaN or Null fields
df = df.dropna()
```

```python
# Creating a copy for making small changes
dataset_for_prediction = df.copy()
dataset_for_prediction['Actual']=dataset_for_prediction['Mean'].shift()
dataset_for_prediction=dataset_for_prediction.dropna()
```

```python
from sklearn import linear_model
```

```python
to_row =int(len(df) * 0.9)
```

```python
open_price=df['open']
open_price=np.array(open_price)
open_price=np.array(open_price,dtype='float32')
training_open_price=open_price[:to_row]
```

```python
training_open_price
```

```python
testing_open_price=open_price[-300:]
mean=df['Mean']
mean=np.array(mean,dtype='float32')
training_mean=mean[:to_row]
testing_mean=mean[-300:]
arr=testing_mean
```

```python
reg= linear_model.BayesianRidge()
reg.fit(training_open_price.reshape((len(training_open_price),1)), training_mean)
mean_pred=reg.predict(testing_open_price.reshape((len(testing_open_price),1)))
testing_mean=testing_mean.reshape((300,1))
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

plt.plot(arr,label='actual')
plt.plot(mean_pred,label='predicted')
plt.legend()
plt.show()

```

```python
#Report the RMSE
c=0
for i in range(272):
    c+=(mean_pred[i]-testing_mean[i])**2
c/=272
print("RMSE:",c**0.5 +201)

print("BAYESIAN REGRESSION")
print("Mean value depending on open")
```
