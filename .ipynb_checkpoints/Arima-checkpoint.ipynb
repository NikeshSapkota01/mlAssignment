```python
# Train test split

to_row =int(len(btc) * 0.9)

training_data = list(btc[0:to_row]['tomorrow'])
testing_data = list(btc[to_row:]['tomorrow'])

to_row
training_data
testing_data
```

```python
# plotting the training and testing data

fig, ax = plt.subplots(figsize=(10, 5))

# plot training data in green
btc[0:to_row]['tomorrow'].plot(ax=ax, color='green')

# plot testing data in red
btc[to_row:]['tomorrow'].plot(ax=ax, color='orange')

# customize the plot
ax.set_title('Bitcoin Training and Testing Data')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.legend(['Training Data', 'Testing Data'])
plt.show()
```

```python
model_predictions = []
n_test_obser = len(testing_data)
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
output = []
training_data_copy = training_data.copy()  # make a copy of the training data to avoid modifying the original
for i in range(n_test_observer):
    model = ARIMA(training_data_copy, order=(4, 1, 0))
    model_fit = model.fit()
    val = model_fit.forecast()
    output.append(val)
    training_data_copy = np.append(training_data_copy, val)
    print(output)
```

```python

for i in range(n_test_observer):
    model = ARIMA(training_data, order = (4,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    model_predictions.append(yhat)
    actual_test_value = testing_data[i]
    training_data.append(actual_test_value)
```

```python
plt.figure(figsize=(15,9))
plt.grid(True)

date_range = btc[to_row+1:].index

plt.plot(date_range, model_predictions[:-1], color = 'orange', marker = 'o', linestyle='dashed', label = 'BTC Predicted Price')
plt.plot(date_range, testing_data[:-1], color = 'green', marker = 'o', linestyle='dashed', label = 'BTC Actual Price')


plt.title('Bitcoin Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
mape = np.mean(np.abs(np.array(model_predictions[:-1]) - np.array(testing_data[:-1]))/np.abs(testing_data[:-1]))
# Mean absoulte percentage error
print('MAPE:' + str(mape))

# Around 3.8% MAPE(Mean absoulte percentage error) implies that the model is about 96.2% 
# accurate in predicting thetest set observations
```
